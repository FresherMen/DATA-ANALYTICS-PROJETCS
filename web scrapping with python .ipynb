```python
import requests
from bs4 import BeautifulSoup
url = "https://www.espncricinfo.com"
```

```python
# Step 1 : Get the HTML

r = requests.get(url)

htmlContent = r.content

print(htmlContent)
```

```python
 # Step 2 : Parse the HTML
soup = BeautifulSoup(htmlContent,"html.parser")
print(soup.prettify)
```

```python
 # Step 3 : HTML Tree Transveral
title = soup.title  
print(title)
#FOR TITLE OF HTML PAGE
```

```python
paras = soup.find_all("p") #used for finding the paragraph
print(paras)
```

```python
anchors = soup.find_all("a") # finding anchors
print(anchors)
```

```python
anchors = soup.find_all("a") 
all_links= set()
# for finding the link inside the anchor tags
for link in anchors:
    if(link.get("href") != "#"):
        linkText = "https://www.espncricinfo.com" + link.get("href")
        all_links.add(link)
        print(linkText)
```

```python
print(soup.find("p")) # get first element  in html page
```

```python
print(soup.find("p")["class"]) # get classes of any element
```

```python
print(soup.find_all("div",class_="ds-p-4")) 
# class finding with its elements
```

```python
print(soup.find("p").get_text()) # getting the text from tags
```

```python
print(soup.get_text()) # getting the text from soup
```
