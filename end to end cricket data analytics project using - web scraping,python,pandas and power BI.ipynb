```python
import pandas as pd
import json
```

# 0) Match Summary processing

#  1) Batting Summary processing

```python
with open("t20_wc_batting_summary.json") as f:
    data = json.load(f)
    
    all_records=[]
    for rec in data:
        all_records.extend(rec["battingSummary"]) # we want single list where al matches are present so we use extend

match_batting = pd.DataFrame(all_records)
match_batting.head(11)
```

# converting dismissal in out and not out

```python
match_batting["out/not out"] = match_batting.dismissal.apply(lambda x : "out" if len(x)>0 else "not out")
match_batting.head(11)
```

# now drop the dismissal as we have got out and not out

```python
match_batting.drop(columns = ["dismissal"],inplace= True)
match_batting.head(11)
```

# now we can remove extra character = e.g (â€ in  Kusal Mendis)

```python
match_batting["batsmanName"] = match_batting["batsmanName"].apply(lambda x: x.replace("â€"," "))
match_batting["batsmanName"] = match_batting["batsmanName"].apply(lambda x : x.replace("(c)"," "))
match_batting.head(11)
```

# Connecting both cleaned data sets with each other :-  
# 1) match_sum and match_batting

# For this we have to make a dictionary in match_sum for unique_id with team   1 and team 2 

```python
unique_ids_dict = {}

for index,row in match_sum.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
        
    unique_ids_dict[key1] = row["unique_id"]
    unique_ids_dict[key2] = row["unique_id"]
    
unique_ids_dict
```

#  Now we have connected both datasets :- As you can see match_id is now in the  match_batting by using mapping function

```python
match_batting["match_id"] = match_batting["match"].map(unique_ids_dict)
match_batting.head(11)
```

```python
match_batting.drop(columns = ["unique_id"],inplace = True)
match_batting.head(11)
```

#  :-> Converting json file in the csv file : match_batting new one 

#            exported batting_summary

```python
match_batting.to_csv("t20_json_files/match_sum_batting_summary.csv",index = False)
```

## 2) Bowling Summary processing

```python
with open("t20_wc_bowling_summary.json") as f:
        data = json.load(f)
        all_records = []
        for rec in data :
            all_records.extend(rec["bowlingSummary"])
            
all_records[:3]       
```

```python
match_bowling = pd.DataFrame(all_records)
print(match_bowling.shape)
match_bowling.head(11)
```

```python
match_bowling["match_id"] = match_bowling["match"].map(unique_ids_dict)
match_bowling.head(11)
```

## exported balling summary 

```python
match_bowling.to_csv("t20_json_files/match_bowling.csv",index = False)
```
